```python
import pandas as pd
import pymysql
```

```python
kospi = pd.read_csv('kospi(18.07.05).csv')
kosdaq = pd.read_csv('kosdaq(18.07.05).csv', error_bad_lines=False)
konex = pd.read_csv('konex(18.07.05).csv', error_bad_lines=False)
```

```python
# MySQL Connection 연결 (DB 연결)
def setting_db():
    conn = pymysql.connect(host="localhost", 
                       port=3306, user="root", 
                       passwd="qlqtmajrwk",
                       db="yejinyDB",
                       charset="utf8")
    cursor = conn.cursor(pymysql.cursors.DictCursor) # 리스트로 받기(tuple은 conn.cursor())
    return conn, cursor
```

```python
# Corp table 만들기
conn, cursor = setting_db()
# 종목코드 crp_cd 법인구분 crp_cls 기업명 crp_nm 업종코드 mrk_cd 업종명 mrk_nm
# 상장주식수 listed(stocks) 자본금 capital 액면가 par(value) 통화 currency 주소 addr
sql = '''CREATE TABLE Corp (crp_cd nvarchar(45) NOT NULL,
                            crp_cls nvarchar(5) NOT NULL,
                            crp_nm nvarchar(150) NOT NULL,
                            mrk_cd nvarchar(45) NOT NULL,
                            mrk_nm nvarchar(150) NOT NULL,
                            listed bigint,
                            capital bigint,
                            par int,
                            currency nvarchar(20),
                            addr nvarchar(100),
                            PRIMARY KEY(crp_cd)
                            );
                            '''
cursor.execute(sql)
conn.commit()
conn.close()
```
