```python
# MySQL Connection 연결 (DB 연결)
def setting_db():
    conn = pymysql.connect(host="localhost", 
                       port=3306, user="root", 
                       passwd="qlqtmajrwk",
                       db="yejinyDB",
                       charset="utf8")
    cursor = conn.cursor(pymysql.cursors.DictCursor) # 리스트로 받기(tuple은 conn.cursor())
    return conn, cursor
```

```python
# stockDB 만들기
conn, cursor = setting_db()
# 종목코드 crp_cd 법인구분 crp_cls 기업명 crp_nm 업종코드 mrk_cd 업종명 mrk_nm
# 상장주식수 listed(stocks) 자본금 capital 액면가 par(value) 통화 currency 주소 addr
sql = '''CREATE TABLE Corp (crp_cd nvarchar(45) NOT NULL,
                            crp_cls nvarchar(5) NOT NULL,
                            crp_nm nvarchar(150) NOT NULL,
                            mrk_cd nvarchar(45) NOT NULL,
                            mrk_nm nvarchar(150) NOT NULL,
                            listed bigint,
                            capital bigint,
                            par int,
                            currency nvarchar(20),
                            addr nvarchar(100),
                            PRIMARY KEY(crp_cd)
                            );
                            '''
cursor.execute(sql)
conn.commit()
conn.close()
```

```python
conn, cursor = setting_db()

for i in range(0, len(kospi)):
    k = kospi.loc[i]
    crp_cd = str(k[1])
    crp_cls = 'Y'
    crp_nm = k[2]
    mrk_cd = str(k[3])
    mrk_nm = k[4]
    listed = k[5].replace(",","")
    capital = k[6].replace(",","")
    par = k[7].replace(",","")
#    listed = int(k[5].replace(",","")) #int
#    capital = float(k[6].replace(",","")) #decimal
#    par = int(k[7].replace(",","")) #int
    currency = k[8]
    addr = k[9]
    sql = "insert into Corp(crp_cd, crp_cls, crp_nm, mrk_cd, mrk_nm, listed, capital, par, currency, addr) select '"+crp_cd+"', '"+crp_cls+"', '"+crp_nm+"', '"+mrk_cd+"', '"+mrk_nm+"', "+listed+", "+capital+", "+par+", '"+currency+"', '"+addr+"';"
    cursor.execute(sql)
    
conn.commit()
conn.close()
```
